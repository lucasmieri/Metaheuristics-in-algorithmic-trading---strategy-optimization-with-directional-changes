```python
import os
import pandas as pd

from src.dc_model_manager import DCModelManager

model_manager=DCModelManager()

for ticker in model_manager.dict_input_parameters.get('b3_tickers', []):
    print(f"Loading data for {ticker}...")
    model_manager.load_ticker_data(ticker)
```

```python
from src.dc.dc_transformer import transform_to_dc_events
from src.dc.dc_graphs import generate_complete_dc_analysis
from pathlib import Path

df = pd.read_parquet(os.path.join(os.getcwd(), 'data', 'bbas3_2025-12-31_2015-12-31.gzip'))

dc_events = transform_to_dc_events(
    df_prices=df,
    threshold=0.02,
    logger=model_manager.logger
)

results = generate_complete_dc_analysis(
    df_dc=dc_events,
    threshold=0.02,
    output_dir=Path('results/dc_analysis_bbas3'),
    thresholds_for_sensitivity=[0.01, 0.02, 0.03, 0.05, 0.10],
    window_days=30,
    logger=model_manager.logger
)
```

```python

%load_ext autoreload
%autoreload 2
from src.strategies.simple_dc_strategy import generate_dc_signals
from src.backtesting.simple_backtest import run_simple_backtest, run_buy_and_hold
from src.backtesting.performance_metrics import calculate_performance_metrics

# 1. Generate signals from DC events
print("="*60)
print("STEP 1: Generating Trading Signals")
print("="*60)

df_signals = generate_dc_signals(
    df_dc=dc_events,
    initial_position='cash',
    logger=model_manager.logger
)

print(f"\nSignals generated:")
print(f"  Buy signals: {(df_signals['signal'] == 'buy').sum()}")
print(f"  Sell signals: {(df_signals['signal'] == 'sell').sum()}")
print(f"  Hold periods: {(df_signals['signal'] == 'hold').sum()}")

# Check first few signals
print(f"\nFirst signals:")
print(df_signals[df_signals['signal'] != 'hold'][['Close', 'event_type', 'signal', 'position']].head(10))

# 2. Run DC Strategy Backtest
print("\n" + "="*60)
print("STEP 2: Running DC Strategy Backtest")
print("="*60)

df_backtest = run_simple_backtest(
    df_signals=df_signals,
    price_column='Close',
    initial_capital=10000.0,
    transaction_cost=0.001,
    logger=model_manager.logger
)

print(f"\nBacktest results:")
print(f"  Initial capital: R$ 10,000.00")
print(f"  Final value: R$ {df_backtest['portfolio_value'].iloc[-1]:,.2f}")
print(f"  Total transaction costs: R$ {df_backtest['trade_cost'].sum():,.2f}")

# 3. Calculate DC Strategy Metrics
print("\n" + "="*60)
print("STEP 3: Calculating DC Strategy Performance")
print("="*60)

dc_metrics = calculate_performance_metrics(
    df_results=df_backtest,
    returns_column='strategy_returns',
    risk_free_rate=0.05,
    logger=model_manager.logger
)

print(f"\nDC Strategy Metrics:")
print(f"  Total Return: {dc_metrics['total_return']:.2%}")
print(f"  Annualized Return: {dc_metrics['annualized_return']:.2%}")
print(f"  Sharpe Ratio: {dc_metrics['sharpe_ratio']:.3f}")
print(f"  Max Drawdown: {dc_metrics['max_drawdown']:.2%}")
print(f"  Volatility: {dc_metrics['volatility']:.2%}")
print(f"  Number of Trades: {dc_metrics['n_trades']}")

# 4. Run Buy & Hold Benchmark
print("\n" + "="*60)
print("STEP 4: Running Buy & Hold Benchmark")
print("="*60)

df_bh = run_buy_and_hold(
    df_prices=dc_events,
    price_column='Close',
    initial_capital=10000.0,
    transaction_cost=0.001,
    logger=model_manager.logger
)

# 5. Calculate Buy & Hold Metrics
bh_metrics = calculate_performance_metrics(
    df_results=df_bh,
    returns_column='bh_returns',
    risk_free_rate=0.05,
    logger=None
)

print(f"\nBuy & Hold Metrics:")
print(f"  Total Return: {bh_metrics['total_return']:.2%}")
print(f"  Annualized Return: {bh_metrics['annualized_return']:.2%}")
print(f"  Sharpe Ratio: {bh_metrics['sharpe_ratio']:.3f}")
print(f"  Max Drawdown: {bh_metrics['max_drawdown']:.2%}")
print(f"  Volatility: {bh_metrics['volatility']:.2%}")

# 6. Comparison
print("\n" + "="*60)
print("STRATEGY COMPARISON")
print("="*60)

comparison = pd.DataFrame({
    'DC Strategy': dc_metrics,
    'Buy & Hold': bh_metrics
}).T

print("\n", comparison)

print("\n" + "="*60)
print("Outperformance:")
print(f"  Return difference: {(dc_metrics['total_return'] - bh_metrics['total_return']):.2%}")
print(f"  Sharpe difference: {(dc_metrics['sharpe_ratio'] - bh_metrics['sharpe_ratio']):.3f}")
print("="*60)
```
